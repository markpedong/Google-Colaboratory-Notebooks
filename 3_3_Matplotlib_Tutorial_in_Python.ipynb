Mathplotlib:

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
print(x)
```

```python
# vertical position on the unit circle
y = np.sin(x) # this assumes that the numbers inputted are on radian
print(y)
```

```python
# horizontal position on the unit circle
z = np.cos(x)
print(z)
```

Plotting the data

```python
# sine wave
plt.plot(x, y)
plt.show()
```

```python
# cosine wave
plt.plot(x, z)
plt.show()
```

```python
# adding a title x-axis & y-axis labels
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'g.')
```

```python
plt.plot(x, y, 'rx')
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0.6,0.2,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people =[100,50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
```

Pie Chart


```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people =[100,50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%') # %1.1f%% is just a format specifier
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10 ,30)
y = np.sin(x)
z= np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x, y, color='g')
ax.scatter(x, z, color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20* np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
